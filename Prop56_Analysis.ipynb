I noticed on this list Thomas F. Steyer and decided to look more into who he is. [Tom Steyer](https://nextgenclimate.org/about-us/tom-steyer/) is co-chair for _Save Lives California_, a coalition to prevent teen smoking and fund cancer research. Below is a dive into Steyer's contributions supporting Prop 56.

```python
steyerContrib = prop56contrib[prop56contrib.contributor_lastname.fillna('NULL').str.contains('STEYER')][['contributor_lastname',\
                                                                                       'contributor_firstname','amount'\
                                                                                       ,'contributor_city']]
```

```python
steyerContrib.reset_index()
```

Steyer contributed over $13 million in support of Prop 56.

```python
steyerContrib.amount.sum()
```

Steyer's contributions amounted to 38% of the total contributions in support of Prop 56.

```python
steyerContrib.amount.sum() / p56support.amount.sum()
```

Steyer's contributions amounted to 14% of the entire sum of contributions, both in support and opposition of Prop 56.

```python
steyerContrib.amount.sum() / prop56contrib.amount.sum()
```
