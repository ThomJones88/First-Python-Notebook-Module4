# Proposition 60 Analysis
I wanted to see contributions that were made towards Proposition 60 or the prop that would require condoms be worn by adult film  actors and actresses.

```python
CondomOppose = props.query("committee_position == 'OPPOSE' and ccdc_prop_id == 79")
```

```python
CondomContribs = contribs.query('calaccess_committee_id in @CondomOppose.calaccess_committee_id')
```

```python
LargeContrib = CondomContribs.loc[CondomContribs.amount.idxmax()]
```

```python
LargeContrib
```

```python
CondomSupport = props.query("committee_position == 'SUPPORT' and ccdc_prop_id == 79")
```

```python
CondomContribsSupport = contribs.query('calaccess_committee_id in @CondomSupport.calaccess_committee_id')
```

```python
LargeSupportContrib = CondomContribsSupport.loc[CondomContribsSupport.amount.idxmax()]
```
